```python
def calc(s1, s2):
    node_tree = dict()
    tree_node = dict()
    min_tree = dict()
    count_tree = 1

    for i in range(len(s1)):
        if node_tree.get(s1[i]) is not None and node_tree.get(s2[i]) is not None:
            tree_no_node_1 = node_tree[s1[i]]
            tree_no_node_2 = node_tree[s2[i]]
            if tree_no_node_1 == tree_no_node_2:
                min_tree[tree_no_node_1] = min(min_tree[tree_no_node_1], min_tree[tree_no_node_2], s1[i], s2[i])
                continue
            elif tree_no_node_1 < tree_no_node_2:
                lst1 = tree_node[tree_no_node_2]
                tree_node[tree_no_node_1].extend(lst1)
                for ele in lst1:
                    node_tree[ele] = tree_no_node_1
                min_tree[tree_no_node_1] = min(min_tree[tree_no_node_1], s1[i], s2[i])
                del tree_node[tree_no_node_2]
            else:
                lst1 = tree_node[tree_no_node_1]
                tree_node[tree_no_node_2].extend(lst1)
                for ele in tree_node[tree_no_node_1]:
                    node_tree[ele] = tree_no_node_2
                min_tree[tree_no_node_2] = min(min_tree[tree_no_node_2], s1[i], s2[i])
                del tree_node[tree_no_node_1] 

        if node_tree.get(s1[i]) is not None and node_tree.get(s2[i]) is None:
            tree_no_node_1 = node_tree[s1[i]]
            node_tree[s2[i]] = tree_no_node_1
            tree_node[tree_no_node_1].append(s2[i])
            min_tree[tree_no_node_1] = min(min_tree[tree_no_node_1], s1[i], s2[i])

        if node_tree.get(s1[i]) is None and node_tree.get(s2[i]) is not None:
            tree_no_node_2 = node_tree[s2[i]]
            node_tree[s1[i]] = tree_no_node_2
            tree_node[tree_no_node_2].append(s1[i])
            min_tree[tree_no_node_2] = min(min_tree[tree_no_node_2], s1[i], s2[i])

        if node_tree.get(s1[i]) is None and node_tree.get(s2[i]) is None:
            node_tree[s1[i]] = node_tree[s2[i]] = count_tree
            tree_node[count_tree] = [s1[i], s2[i]]
            min_tree[count_tree] = min(s1[i], s2[i])
            count_tree += 1
            
    return node_tree, min_tree
            
def replaced(s1, s2, s3):
    node_tree, min_tree = calc(s1, s2)
    ans = ""
    for ele in s3:
        if node_tree.get(ele) is not None:
            ans += min_tree[node_tree[ele]]
        else : ans += ele
    return ans
        
if __name__ == "__main__":
    s1 = input()
    s2 = input()
    s3 = input()
    print(replaced(s1, s2, s3))
```
