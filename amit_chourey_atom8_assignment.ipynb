```python
def res(s1, s2, s3):
    dict1 = {}
    for i in range(len(s1)):
        if dict1.get(s1[i]) is None:
            dict1[s1[i]] = set()
        dict1[s1[i]].update({s2[i]})
        if dict1.get(s2[i]) is None:
            dict1[s2[i]] = set()
        dict1[s2[i]].update({s1[i]})

    while True:
        dict2 = {}
        s = 0
        for key, val in dict1.items():
            for ele in val:
                if dict2.get(key) is None:
                    dict2[key] = set()
                dict2[key].update((dict1[key]))
                n = len(dict2[key])
                dict2[key].update((dict1[ele]))
                s = s + len(dict2[key]) - n
        dict1 = dict2
        if s == 0 :
            break

    final = {}
    for key, val in dict1.items():
        final[key] = sorted(val)[0]
    s4 = ""
    for ele in s3:
        s4 += final[ele]
    return s4
```
